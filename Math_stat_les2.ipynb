1. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

Используем уравнение Бернулли, поскольку количество экспериментов не большое, а вероятность события высокая.<br>
$P_{100}=C^{85}_{100}*0.8^{85}*0.2^{15}=\frac{100!}{85!15!}*0.8^{85}*0.2^{15}=0.048$ Т.е. 4.8%.<br>
Странно, по логике 80 попаданий из 100 должны быть равны 100%. Неужели 5 дополнительных выстрелов так сильно снижают вероятность?

2. Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?

Одна лампочка перегорит в первый день с вероятностью $0.0004*5000=2$% Следовательно, ни одна не перегорит с вероятностью 98%<br>
Вторая перегорит с вероятностью $0.0004*4999=1.9996$ Следовательно 2 перегорит с вероятностью 3.9996%

3. Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?

Используем уравнение Бернулли, поскольку количество экспериментов не большое, а вероятность события высокая.<br>
$P_{144}=C^{70}_{144}*0.5^{70}*0.5^{74}=\frac{144!}{70!74!}*0.5^{70}*0.5^{74}=0.0628$ Т.е. 6.28%.<br>

4. В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что все мячи белые? Какова вероятность того, что ровно два мяча белые? Какова вероятность того, что хотя бы один мяч белый?

Количество комбинаций белых в 1-ом ящике:$C^2_7=\frac{7!}{2!5!}=21$<br>
Общее кол-во комбинаций:$C^2_{10}=\frac{10!}{2!8!}=45$<br>
Вероятность 2-х белых в первом ящике:$\frac{21}{45}=0.46$ Или 46%

Количество комбинаций белых во 2-ом ящике:$C^2_9=\frac{9!}{2!7!}=36$<br>
Общее кол-во комбинаций:$C^2_{11}=\frac{11!}{2!9!}=55$<br>
Вероятность 2-х белых во втором ящике:$\frac{36}{55}=0.65$ Или 65%

События по обоим ящикам независимые и сумма вероятностей здесь не работает. По логике, общая вероятность будет среднее арифметическое.<br> Т.е. $\frac{65+46}{2}=55.5%$ Прикрутить Бернулли к решению так и не смог.
