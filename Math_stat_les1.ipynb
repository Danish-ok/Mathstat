1. Из колоды в 52 карты извлекаются случайным образом 4 карты.<br>
a) Найти вероятность того, что все карты – крести.<br> 
б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

a) Мы имеем 13 карт крести в колоде. Они могут выпасть $\frac{13!}{4!(13-4)!}=\frac{13!}{4!9!}=715$ вариантов.<br>
Общее количество вариантов 4-х карт в колоде из 52 карт: $\frac{52!}{4!48!}= 270725$<br>
Вероятность 4-х крести: $\frac{715}{270725}=0,002641$ Логика вроде правильная, но смущает низкая вероятность...

б) Один туз выпадает $C^1_4=\frac{4!}{1!(4-1)!}=4 $ количество раз.<br>
Вариантов выпадения 3-х карт $C^3_{48}=\frac{48!}{3!45!}=17296$ без учета тузов.<br>
Количество выпадений одного туза с 3-мя картами 17296*4=69184.

Два туза выпадают $C^2_4=\frac{4!}{2!2!}=6$ количество раз.<br>
Вариантов выпадения 2-х карт $C^2_{48}=\frac{48!}{2!46!}=1128$ без учета тузов.<br>
Количество выпадений двух тузов с 2-мя картами 1128*6=6768.

Три туза выпадают $C^3_4=\frac{4!}{3!1!}=4$ количество раз.<br>
Вариантов выпадения 1-й карты $C^1_{48}=48$ без учета тузов.<br>
Количество выпадений трех тузов с 1-й картой 4*48=192.

Четыре туза выпадают 1 раз.<br>
Общее количество вариантов раздачи $C^4_52=\frac{52!}{4!48!}=270725$<br>
Вероятность, что среди 4-х карт окажется хотя бы один туз: $\frac{1+192+6768+69184}{270725}=0,281263$ Т.е. около 28%

2. На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

Учитывая, что кнопки нажимают одновременно, видимо порядок цифр роли не играет. <br>
Тогда, общее количество комбинаций: $C^3_{10}=\frac{10!}{3!7!}=120$<br>
Вероятность удадать комбинацию с первой попытки: $\frac{1}{120}=0,008$ Т.е. около 0,8%

3. В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

Общее количество вариантов извлечения деталей: $C^3_{15}=\frac{15!}{3!12!}=455$<br>
Количество вариантов извлечения окрашенных деталей: $C^3_{9}=\frac{9!}{3!6!}=84$<br>
Вероятность достать три окрашенные: $\frac{84}{455}=0,185$ Т.е. около 18,5%

4. В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

Количество вариантов пар билетов: $C^2_{100}=\frac{100!}{2!98!}=4950$<br>
2 из 2 можно купить только одним способом. Следовательно, вероятность купить 2 выигрышных билета $\frac{1}{4950}=0,0002$ Т.е. 0,02%.
